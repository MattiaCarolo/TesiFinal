```python
from itertools import chain

# Specify the indices of the wanted tracks for each group
group1_indices = [0, 1, 2]  # Example indices for the first group
group2_indices = [3, 4, 5]  # Example indices for the second group

# Filter notes for the first group of tracks
group1_notes = chain(*[
    midi.instruments[i].notes for i in group1_indices
])

# Filter notes for the second group of tracks
group2_notes = chain(*[
    midi.instruments[i].notes for i in group2_indices
])

# Now you have two separate iterables for each group of notes
```
